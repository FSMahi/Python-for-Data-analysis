**Analyzing a data from data frames**

What is the total number of reported cases and deaths related to Covid-19 in Italy?

```python
total_cases= covid_df.new_cases.sum()
total_deaths= covid_df.new_deaths.sum()
```

```python
print('The number of reported cases is {} and the number of reported deaths is {}.'.format(total_cases, total_deaths))
```

```python
print(f'The number of reported cases is {total_cases} and the number of reported deaths is {total_deaths}.')
```


What is the overall death rate (ratio of reported deaths to reported cases)?

```python
death_rate= covid_df.new_deaths.sum() / covid_df.new_cases.sum()
```

```python
print("The overall reported death rate in Italy is {:.2f} %.".format(death_rate*100))
```

What is the overall number of tests conducted? A total of 935310 tests were conducted before daily test numbers were being reported.

```python
covid_df.new_tests.sum()
```

```python
initial_test = 935310
total_test = initial_test + covid_df.new_tests.sum()
```

```python
total_test
```

what fraction of test returned a positive result?

```python
positive_rate = total_cases/total_test
print('{:.2f}% of tests in Italy led to a positive diagnosis.'.format(positive_rate*100))
```

# Grouping and aggregation

```python
covid_df.groupby('month')
```

```python
covid_month_df = covid_df.groupby('month')[['new_cases', 'new_deaths','new_tests']].sum()
covid_month_df
```

1st wave is in may month as the new cases and deaths rate are high this time

```python
covid_df_weekday = covid_df.groupby('weekday')[['new_cases', 'new_deaths','new_tests']].mean()
covid_df_weekday
```

```python
covid_df_weekday_mean = covid_df.groupby('weekday')[['new_cases', 'new_deaths','new_tests']].mean()
covid_df_weekday_mean
```

```python
covid_df['total_cases']= covid_df.new_cases.cumsum()
covid_df['total_deaths']= covid_df.new_deaths.cumsum()
covid_df['total_tests']= covid_df.new_tests.cumsum() + initial_test
covid_df
```

```python
locations_df= pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location == "Italy"]
```

```python
locations_df[locations_df.location == 'Italy'].count()
```

```python
covid_df['location']= 'Italy'
```

```python
covid_df
```

```python
merged_df = covid_df.merge(locations_df, on='location')
```

```python
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
merged_df
```

```python
result_df = merged_df [[ 'date',
                         'new_cases',
                         'total_cases',
                         'new_deaths',
                         'total_deaths',
                         'new_tests',
                         'total_tests',
                         'cases_per_million',
                         'deaths_per_million',
                         'tests_per_million']]
```

```python
result_df
```

```python
result_df.to_csv('results.csv', index=None)
```
