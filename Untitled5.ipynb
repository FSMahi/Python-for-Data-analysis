# Modified Z Score

```python
df_m = pd.read_csv('movie_revenues.csv')
df_m.head()
```

```python
df_m.describe()
```

```python
df_m.revenue.describe()
```

```python
df_m['revenue_mln'] = df_m['revenue'].apply(lambda x: x/1000000)
df_m.revenue_mln.describe()
```

```python
_, mean, std, *_ = df_m.revenue_mln.describe()
```

```python
mean,std
```

# Outlier detection using Z score

```python
def get_z_score(value, mean, std):
    return (value - mean)/std
```

```python
df_m['z_score'] = df_m.revenue_mln.apply(lambda x: get_z_score(x, mean, std))
df_m.head(3)
```

**General guideline is to treat anything that has a z score of 3 or more as an outlier**

```python
df_m[df_m.z_score>3]
```

# **Outlier detection using modified Z score**

So we got only 1 outlier using simple Z score. Based on the kind of analysis we are doing sometimes it might be better to use modified Z score as it will return more outliers.

```python
def get_mad(s):
    median= np.median(s)
    diff= abs(s- median)
    MAD= np.median(diff)
    return MAD
```

```python
MAD = get_mad(df_m.revenue_mln)
median = np.median(df_m.revenue_mln)
MAD, median
```

**General guideline for modified Z score is to use 3.5 as a thresold. i.e. anything that has a mod z score of 3.5 or more is an outlier**

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

**Test z score function for a sample data point**

```python
get_modified_z_score(2711, median, MAD)
```

```python
df_m['mod_z_score'] = df_m.revenue_mln.apply(lambda x: get_modified_z_score(x, median, MAD))
df_m.head(3)
```

```python
df_m[df_m.mod_z_score> 3.5]
```
